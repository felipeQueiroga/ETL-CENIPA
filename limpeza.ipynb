# Limpeza das colunas

```python
df.loc[df.ocorrencia_uf == '***', ['ocorrencia_uf']] = pd.NA
df.loc[df.ocorrencia_aerodromo == '****', ['ocorrencia_aerodromo']] = pd.NA
df.loc[df.ocorrencia_aerodromo == '*****', ['ocorrencia_aerodromo']] = pd.NA
df.loc[df.ocorrencia_aerodromo == '###!', ['ocorrencia_aerodromo']] = pd.NA
df.loc[df.ocorrencia_aerodromo == '####', ['ocorrencia_aerodromo']] = pd.NA
df.loc[df.ocorrencia_hora == 'NULL', ['ocorrencia_hora']] = pd.NA
#trata a tabela toda
df.replace(['**','###!','####','****','*****','NULL'], pd.NA, inplace=True)
```
